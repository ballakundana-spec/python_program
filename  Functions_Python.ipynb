```python
#user defined functions
def summery():
    print("def is keyword to start the functions")
    print("summery is a function name")
    print("arguments or keywords are optional to pass the data to inside the function")
    print("calling the function then run all statements inside the functions")
summery()    
```

```python
#write a program to print addition of two numbers using functions
def addition():
    a=int(input("please enter first number:"))
    b=int(input("please enter second number:"))
    result=a+b
    print(result)
    return result
addition()
```
