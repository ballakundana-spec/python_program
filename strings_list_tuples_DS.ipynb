```python
#acccessing characters from string
#two ways 1.forward index 2.backward index(negetive index)
#forward index - left to right (0,1,2,...n-1)
#backward index - right to left (-1,-2,-3,...-n)
info="welcome to tia"
print(info)
print(info[0])
print(info[11])
print(info[-3])
```

```python
#slicing - accessing group of characters from string - [start_index:end_index-1]
info="welcome to tia"
print(info)
print(info[11:14])
print(info[-3:])
```

# lists in data strutures
1. list is collection of elements enclosed by []
2. its allowed duplicate elements
3. each element seperated by ,
4. list is mutable - editable

```python
# create list
names=["shiva","vikas","divya","keshav"]
ages=[25,26,24,27,27,25]
print(ages)
print(names)
print(type(names))
print(type(ages))
```

```python
#concatenation of list
a1=[10,20,30,40,50]
a2=[10,25,12,56,13,58]
a3=a1+a2
print(a3)
```

```python
#repeat list
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info*5)
```

```python
#accessing elements from list
#index bases
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info[0])
print(info[-6])
```

```python
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info[2:5])
print(info[0:2])
print(info[5:7])
print(info[2:7])
```

```python
# methods or functions available in list -without underscore
print(dir(list))
```

```python
# methods ith discription
print(help(list))
```

```python
#append
names=["shiva","vikram","divya"]
names.append("chandu")
print(names)
```

```python
#entend()
names=["shiva","vikram","divya"]
names.extend(["chandu","vikas"])
print(names)
```

```python
#insert()
names=["shiva","vikram","divya"]
names.insert(1,"vikas")
print(names)
```

```python
#pop()
numbers=[10,20,30,40,50]
print(numbers)
numbers.pop(2)
print(numbers)
```

```python
#remove()
ages=[25,26,27,28,29]
print(ages)
ages.remove(28)
print(ages)
```

```python
#sorting techniques sort() and reverse
numbers=[6,5,8,1,3,4,0,7,2,9]
print(numbers)
numbers.sort()
print(numbers)
```

```python
numbers=[6,5,8,1,3,4,0,7,2,9]
print(numbers)
numbers.sort(reverse=True)
print(numbers)
```

```python
numbers=[6,5,8,1,3,4,0,7,2,9]
print(numbers)
numbers.sort(reverse=False)
print(numbers)
```

# tuples in data structures
1. tuple is a collection of elements enclosed by round brackets ()
2. its allowed duplicate
3. each element seperated by comma ,
4. tuple is a immutable - not editable

```python
#create tuple
sal=(1000,2000,3000,45000)
print(sal)
print(type(sal))
```

```python
#concatenation of tuple
a1=(1,2,3,4,5)
a2=(4,5,6,7,8)
a3=a1+a2
print(a3)
```

```python
#repeat of tuple
names=("nakshatra","priya","jetavya")
print(names*5)
```

```python
#accessing elements from elements and slicing(acccessing group of elements)
info=("priya","nakshatra","jetavya",50,100,500,1000)
print(info)
print(info[0])
print(info[0:3])
print(info[3:7])
```

```python
#methods or functions
print(dir(tuple))
```

```python
ages=(10,20,10,20,25,23,24,26)
print(ages)
print(ages.count(10))
print(ages.count(23))
```

```python
ages=(10,20,10,20,25,23,24,26)
print(ages)
print(ages.index(24))
```
