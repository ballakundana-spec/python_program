```python
====conditional statements====
1. write a python program to check wheather a number is positive, negative or zero
2. write a python program to check if a number is even or odd.
3. write a python program to the greatest of two numbers using if-else.
4. write a python program to find the smallest of two number using nested if statements.
5. write a python program to check if a year is a leaf year or note. 
```

```python
num= float(input("enter a number:"))
if num>0:
    print("It is a positive number")
elif num<0:
    print("It is a negative number")
else:
    print("It is a zero")
```

```python
num= int(input("enter a number:"))
if num%2==0:
    print("It is a even number:")
else:
    print("it is a odd number")
```

```python
a= float(input("enter the first number:"))
b= float(input("enter the second number:"))
if a>b:
    print("the greatest number is:",a)
else:
    print("the greatest number is:",b)
```

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

if num1 > num2:
    print(f"The greatest number is {num1}.")
elif num2 > num1:
    print(f"The greatest number is {num2}.")
else:
    print("Both numbers are equal.")
```

```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
sum=0

for i in range (0,21):
    sum +=i

print("the sum of first 20 natural numbers is",sum)
```

```python
num= int(input("please enter number  here:"))
factorial =1

for i in range(1, num+1):
    factorial *=i
print("the factorial of",num, "is:", factorial)
```

```python
num= int(input("please enter a number here:"))


n=abs(num)
count=0

if n==0:
    count=1

else:
    while n>0:
        n//=10
        count +=1
print("number of digits in", num,"is:", count)
```
