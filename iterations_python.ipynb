````python
# write a program to print names using for loop
names=["eren","mikasa","armin","levi","hange"]
for n in names:
    print("my name is",n)
````

````python
# write a program to print even numbers in between 1 to 20 using for loop
for i in range(1,21):
    if (i%2==0):
         print(i,end=",")
````

````python
# write a program to print odd numbers in between 50 to 100 using for loop
for i in range(50,101):
    if (i%2!=0):
         print(i,end=",")
````

````python
# write a programme to find the vowels from string using loop
name=str(input("please enter your name here:"))
print(name)
for char in name:
    if (char in ["a","e","i","o","u"]):
        print(char)
    
````

````python
#write a program to find the non vowels from string using for loop
name=str(input("please enter your name here:"))
print(name)
for char in name:
    if (char not in ["a","e","i","o","u"]):
        print(char,end=",")
````

````python
# write a program to find the sum of all numbers in between 1 to 20 using for loop
result=0
for i in range(1,21):
    result=result+i
print("sum of all numbers in between 1 to 20 is",result)
````

````python
# write a program to find the sum of all odd numbers in between 1 to 1000 using for loop
result=0
for i in range(1,1001):
    if i%2!=0:
        result=result+i
print("sum of all  odd numbers in between 1 to 1000 is",result)
````

````python
# write a program to find the product of all even numbers in between 500 to 600 using for loop
result=1
for i in range(500,601):
    if i%2==0:
        result=result*i
print("product of all even numbers in between 500 to 600 is",result)
````

````python
# write a program to find the numbers in between 1 to 20 using while loop
sn=1
en=20
while (sn<=en):
    print(sn,end=",")
    sn=sn+1
````

````python
# write a program to find the odd numbers in between 100 to 200 using while loop
sn=100
en=200
while (sn<=en):
       if (sn%2!=0):
           print(sn,end=",")
       sn=sn+1
````
