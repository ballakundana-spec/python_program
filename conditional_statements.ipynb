````python
"""
write a program to find out the grade based on students marks
1. if marks is b/w 80 to 100 grade A
2. if marks is b/w 60 to 79 grade B
3. if marks is b/w 40 to 59 grade c
4. if marks is b/w 25 to 39 grade D
5. if marks is b/w 0 to 24 grade fail
6. if marks other than above output should be please check valid marks
"""
marks = int(input("please enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks,"A grade")
elif (marks>=60 and marks<=79):
    print(marks,"B grade")
elif (marks>=40 and marks<=59):
    print(marks,"C grade")
elif (marks>=25 and marks<=39):
    print(marks,"D grade")
elif (marks>=0 and marks<=24):
    print(marks,"FAIL grade")
else:
    print(marks,"invalid")
````

````python
n1=int(input("please enter 1st number here:"))
n2=int(input("please enter 2nd number here:"))
n3=int(input("please enter 3rd number here:"))
if (n1>=n2 and n1>=n3):
    print(n1," is greater")
elif (n2>=n1 and n2>=n3):
    print(n2," is greater")
else:
    print(n3," is greater") 
````
