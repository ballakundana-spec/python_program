# file handling in python
1. file is collection of information stored in memory
2. files are 2 types- `1.test files` and `2.binary files`

# file types
1. create files - w mode
2. write data to files - w mode
3. read data from files - r mode
4. append data to files - a mode

```python
# write a program to create text file using python script
create_file=open("profile.txt","w")
print("file created successfully and verifies in the home dictionary")
create_file.close()
```

```python
# write data to files using python script
write_data=open("profile.txt","w")
info="""welcome to texas international academy
learning python with data analysis
python with machine learning
python with artificial intelligence
"""
write_data.write(info)
print("successfully data write files and open profile.txt file see the information")
write_data.close()
```

```python
# read data from file using python script
read_file=open("profile.txt","r")
output=read_file.read()
print(output)
read_file.close()
```

```python
# append data to files using python script
append_file=open("profile.txt","a")
data="\nwelcome to full stack development program using python"
append_file.write(data)
print("update data to file and verify the profile.txt file")
append_file.close()
```
